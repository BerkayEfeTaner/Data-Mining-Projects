```python
plt.plot([1,2,3],[7,3,9]) # (1,7),(2,3),(3,9) noktalarını bireştirri
```

```python
x=np.linspace(-np.pi,np.pi,20) # -pi ile pi arasında 20 adet değer
plt.plot(x,np.sin(x))
```

# Grafiklere özellik ekleme
label özelliği verilerin neler olduğunu gösterir.label özelliğinin grafik içerisinde gösterilmesi için plt.legend() fonk ekle

```python
plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend()
```

```python
plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend(loc='best')# upper left upper right. grafiği analiz eder boş yere yazdırır.
```

```python
plt.plot(x,np.sin(x),label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend(loc='upper center')
```

```python
# Ya da aynı figür üzerinde şekil çizmek istersek
plt.plot(x,np.sin(x),x,np.cos(x),x,x**2)
```

```python
#x label,y label ve title ları yerine yazdırma. Mesela yukarıya x,y grafiği aşağıya x egerleri yazıldı
plt.plot(x,np.sin(x),x,np.cos(x),x,x**2)
plt.xlabel("x degerleri")
plt.ylabel("y degerleri")
plt.title("x,y grafigi")
```

```python
# etiketlerin fontlarını eğiştirme
plt.plot(x,np.sin(x),x,np.cos(x),x,x**2)
plt.xlabel("x degerleri")
plt.ylabel("y degerleri")
plt.title("x,y grafigi",
          fontdict={
              'family':'serif',
              'color':'red',
              'size':20},
          loc='right')
```

```python
#grafikte ızgaraları(grid)gösterme
plt.plot(x,np.sin(x),x,np.cos(x),x,x**2);
plt.grid(True)# axis='x'sadece x ekseninde ızgara gösterir.Bir şey demezsek false gelir
```

```python
plt.plot(x,np.sin(x),x,np.cos(x),x,x**2,linewidth=5);#linewidth çizgilerin kalınlıklarını ayarlar
plt.grid()
```

```python
plt.plot(x,np.sin(x),linestyle='dashed',label='sin') # çizgi şekilleri
plt.plot(x,np.cos(x),linestyle='--',label='cos')
plt.plot(x,x**2,label='kare',linestyle='-.')
plt.grid()
plt.legend()
```

# Çoklu figürler il matplotlib grafiği

```python
#basit matplotlib grafiği
fig,ax=plt.subplots()# tek eksenden oluşan bir figür,fig kaç cm e kaç cm
ax.plot([1,2,3,4],[1,4,2,3])# 
```

```python
fig=plt.figure()# Eksenin olmadığı bir figür
fig,ax=plt.subplots()# tek eksenden oluan bir figür
fig,axs=plt.subplots(2,2)# 2x2 li bir grafik
```

```python
fig,ax=plt.subplots()
ax.plot(x,np.sin(x),label='sinüs')
ax.plot(x,np.cos(x),label='cosinüs')
ax.plot(x,x**2,label='kare')
ax.set_xlabel('x axisi')
ax.set_ylabel('y axisi')
ax.set_title("x ve y grafiği")
ax.legend()
```

```python
fig,ax=plt.subplots(1,2)# bir satır iki sütundan ooluşan iki grafik
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x))
```

```python
fig,ax=plt.subplots(2,2) # 2x2 grafik
ax[0][0].plot(x,np.sin(x))
ax[0][1].plot(x,np.cos(x))
ax[1][0].plot(x,x**2)
ax[1][1].plot(x,x**3)
```

#Izgaralar ile subplot çizdirme

```python
for i in range(1,5):
    plt.subplot(2,2,i)# 2 satır 2. sütunun i. ekseni
    plt.text(0.5,0.5,str(i))
```

```python
#plt.subplot2grid() fonksiyonu
a1=plt.subplot2grid((3,3),(0,0),colspan=2)# 3x3 lük zıgarada 1. satır birinci sütunda iki sütun 
a2=plt.subplot2grid((3,3),(0,2),rowspan=3)
a3=plt.subplot2grid((3,3),(1,0),rowspan=2,colspan=2)
x=np.arange(1,10)
a2.plot(x,x*x)
a2.set_title('square')
a1.plot(x,np.exp(x))
a1.set_title('exp')
a3.plot(x,np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

```python
# grafiklere aha fazla eleman ekleme
# metin ekleme
plt.plot([1,2,3])
plt.text(1,2,"x=1,y=2",bbox={'facecolor':'red','alpha':0.1})
plt.grid()
```
